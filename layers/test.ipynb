```python
from lorentz import Lorentz
import torch
```

```python
man = Lorentz(k=1.0)
xs = torch.randn(2, 3, 4, 5)  # Example tensor of shape (2, 3, 4, 5)
time = torch.sqrt(1 + torch.sum(xs ** 2, dim=-1, keepdim=True))  # Shape (2, 3, 4, 1)
xs = torch.cat([time, xs], dim=-1)  # Shape (2, 3, 4, 6)
```

```python
man.lorentz_midpoint(xs)
```

```python
man.distL(xs, xs)
```
